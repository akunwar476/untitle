```python
# creating anew column 'RAM1' and assigning all values of 'RAM' column
data['RAM1']=data['RAM']
data.head()
```

```python
# creating a new column 'RAM2,RAM3,RAM4' and assigning all values of 'RAM' column
data['RAM2']=data['RAM']
data['RAM3']=data['RAM']
data['RAM4']=data['RAM']
data.head()
```

```python
# delete a single column
#Here i will delete column'RAM1'
del data['RAM1']
data.head()
```

```python
# drp multiple column at a time
# here i will drop column RAM2,RAM3,RAM4
# finally all modified stored in data object
data= data.drop(data.columns[[9,10,11]],axis=1)
data.head()
```

```python
#Remove the ₹ symbol from the 'Price' column
data['Price1']= data['Price'].apply(lambda x: x.replace('₹',''))
data.head()
```

```python
#Remove the phrase 'Front Camera' and 'MP' in every attribute of the column 'Front Camera'
data['Front Camera1']=data['Front Camera'].apply(lambda x: x.split('Front Camera')[0])
data.head()
```

```python
# suppose i want to delte MP too from Front Camera
data['Front Camera2']=data['Front Camera'].apply(lambda x: x.split('MP')[0])
data.head()
```

```python
# Remove all the phrases after 'mAh' in the battery column including phrase 'mAh'
data['Battery1']= data['Battery'].apply(lambda x:x.split('mAh')[0])
data.head()
```

```python
#Remove all the phrases after '(' in the Display column
data['Display'] =data['Display'].apply(lambda x: x.split('cm')[0])
data.head()
```

```python
data['Name1']=data['Name'].apply(lambda x: x.split(',')[0])+')'
data.head()
```

```python
# Remove 'RAM' term in the RAM column
data['RAM1']= data['RAM'].apply(lambda x: x.split('RAM')[0])
data.head()
```

```python
#Remove 'ROM' term in the ROM column
data['ROM1']= data['Internal storage'].apply(lambda x: x.split('ROM')[0])
data.head()
```
